## Task: Price Range Distribution

### Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

### Calculate the percentage of restaurants in each price range category.

### Counting the number of occurrences for each price range

```python
pricecount = df['Price range'].value_counts()
pricecount
```

### Ploting the result

```python
sns.barplot(x=pricecount.index, y=pricecount.values, palette='Blues')
plt.title('Distribution of Price Ranges', fontsize=14)
plt.xlabel('Price Range', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.show()
```

### Calculating the percentage distribution of price ranges


```python
percentage = df['Price range'].value_counts(normalize=True) * 100
```

### Piechart for data visualization

```python
plt.figure(figsize=(6,5))
plt.pie(percentage,labels=percentage.index,autopct='%1.1f%%')
plt.title("Pie chart for price range distribution")
plt.show()
```
