## Level 1- Task1: Top Cuisines

### Determine the top three most common cuisines in the dataset.

### Calculate the percentage of restaurants that serve each of the top cuisines.



### Splitting the cuisines in each row and counting its frequency



```python
cuisinecount = df["Cuisines"].str.split(",").explode("Cuisines").str.strip().value_counts()
print(cuisinecount)
```

### Finding the top 3 most common cuisines



```python
top3cuisines = cuisinecount.head(3)
top3cuisines
```

### Ploting the result

```python
sns.barplot(x=top3cuisines.index, y=top3cuisines.values,palette='Purples')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()
```

### percentage of restaurants serving each top cuisine


```python
totalrestaurants = df['Restaurant ID'].nunique()  
topcuisinepercentages = (top3cuisines / totalrestaurants) * 100

print("Top 3 Cuisines and their percentages:")
print(topcuisinepercentages)
```

### Ploting the result

```python
sns.barplot(x=topcuisinepercentages.index, y=topcuisinepercentages.values,palette='Reds')
plt.xlabel('Cuisine')
plt.ylabel('Percentage (%)')
plt.show()
```
