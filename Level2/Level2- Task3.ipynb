## Task: Geographic Analysis
### Plot the locations of restaurants on a map using longitude and latitude coordinates.
### Identify any patterns or clusters of restaurants in specific areas.


 ### Importing the required libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the data set


```python
df=pd.read_excel(r"C:\Users\Lenovo\Downloads\cleaned_data_cognifyz.xlsx")
df
```

### Ploting Restaurant Locations as a Scatter Plot

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Longitude', y='Latitude', data=df, hue='City', palette='viridis', s=50, edgecolor='black')
plt.title('Restaurant Locations on Map', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
