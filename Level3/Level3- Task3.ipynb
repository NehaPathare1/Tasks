## Task: Price Range vs. Online Delivery and Table Booking
#### Analyze if there is a relationship between the price range and the availability of online delivery and table booking.
#### Determine if higher-priced restaurants are more likely to offer these services.


 ### Importing the required libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the data set

```python
df=pd.read_excel(r"C:\Users\Lenovo\Downloads\cleaned_data_cognifyz.xlsx")
df
```

```python
df.columns
```

### Grouping data by Price Range and calculating the count of restaurants with online delivery

```python
d=df[['Price range','Has Online delivery']][df['Has Online delivery']=="Yes"].groupby('Has Online delivery').count()
d
```

### Grouping data by Price Range and calculating the count of restaurants with table booking


```python
b=df[['Price range','Has Table booking']][df['Has Table booking']=="Yes"].groupby('Has Table booking').count()
b
```

 ### Visualizing Price Range vs Online Delivery

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Price range', hue='Has Online delivery', data=df)
plt.title('Price Range vs Online Delivery')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

 ### Visualizing Price Range vs Table Booking


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Price range', hue='Has Table booking', data=df)
plt.title('Price Range vs Table Booking')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

### Insights

```python
# With increase in price range there is an increase in the online delivery rate but is opposite in the case of Table booking facility.
```
