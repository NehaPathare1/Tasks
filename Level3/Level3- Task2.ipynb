### Task: Votes Analysis
#### Identify the restaurants with the highest and lowest number of votes.
#### Analyze if there is a correlation between the number of votes and the rating of a restaurant.

 ### Importing the required libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the data set

```python
df=pd.read_excel(r"C:\Users\Lenovo\Downloads\cleaned_data_cognifyz.xlsx")
df
```

```python
df.columns
```

### Restaurant with the lowest votes 

```python
df[['Restaurant Name','Votes']].groupby('Restaurant Name').count().sort_values(by = 'Votes',ascending=True)
```

### Restaurant with the Highest votes 

```python
df[['Restaurant Name','Votes']].groupby('Restaurant Name').count().sort_values(by = 'Votes',ascending=False).head(1)
```

### Correlation Analysis

```python
# Correlation between 'Votes' and 'Aggregate rating'
correlation = df['Votes'].corr(df['Aggregate rating'])
correlation
```

### Visualize the relationship using a scatter plot

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Votes', y='Aggregate rating', color='green')
plt.title('Votes vs Aggregate Rating', fontsize=16)
plt.xlabel('Number of Votes', fontsize=12)
plt.ylabel('Aggregate Rating', fontsize=12)
plt.show()
```
